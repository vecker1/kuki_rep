```python
import pandas as pd

import seaborn as sns
sns.set(color_codes=True)

import matplotlib.pyplot as plt

data = pd.read_csv("zootest.csv")

data
```

```python
conteo = data['hair_boolean'].value_counts()

ax=conteo.plot(kind='bar', color=['blue', 'green'])

plt.title('Cantidad de animales que tienen pelo')
plt.xlabel('Booleano')
plt.ylabel('Cantidad')
plt.xticks([0, 1], ['no', 'si'], rotation=0)  # Etiquetas del eje x

for i, v in enumerate(conteo):
    ax.text(i, v + 0.1, str(v), color='black', fontweight='bold', ha='center')

plt.show()
```
